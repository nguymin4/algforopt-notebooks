```julia
import LinearAlgebra
import Plots as plt

include("../lib/test_functions.jl")
import .TestFunctions: ackley
```

```julia
plt.plotlyjs()

f = ackley()

xdomain = LinRange(-5, 5, 100)
ydomain = LinRange(-5, 5, 100)

surface = plt.surface(xdomain, ydomain, (x, y) -> f([x, y]); opacity=0.9, colorbar=false, fill=(true, plt.cgrad(:vik)))
contour = plt.contour(xdomain, ydomain, (x, y) -> f([x, y]); colorbar=false, fill=(true, plt.cgrad(:vik)))
plt.plot(surface, contour, layout=(1, 2), size=(960, 480))
```
